### Artificial Neural Network

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import operator
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import metrics
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import multilabel_confusion_matrix, accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
```

```python
# Create a path
path = r'C:\Users\Asus-Pc\Desktop\CareerFoundry\Machine Learning with Python'
```

```python
# Import [DATASET]_scaled.csv
df_scaled = pd.read_csv(os.path.join(path, '[DATASET]_scaled.csv'), index_col = False)
df_scaled
```

```python
# Import weather dataset
df_weather = pd.read_csv(os.path.join(path, 'Weather_dataset.csv'), index_col = False)
df_weather
```

```python
df_scaled.columns
```

```python
df_weather.columns
```

```python
#Drop columns regarding Roma, Tours and Gdansk from scaled dataset

df_reduced = df_scaled.drop(columns = ['Unnamed: 0','GDANSK_temp_mean', 'GDANSK_temp_min', 'GDANSK_temp_max', 'TOURS_temp_mean', 'TOURS_temp_min',
                            'TOURS_temp_max', 'ROMA_temp_mean'])
```

```python
df_reduced.columns
```

```python
#Create an 'X' matrix by dropping the unneeded columns.
X = df_reduced.drop(['DATE', 'MONTH'], axis=1)
y = df_weather.drop(['DATE'], axis=1)
```

```python
print(X.shape)
print(y.shape)
```

```python
#Split data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Create the ANN
#hidden_layer_sizes has up to three layers, each with a number of nodes. So (5, 5) is two hidden layers with 5 nodes each,
#and (100, 50, 25) is three hidden layers with 100, 50, and 25 nodes.
mlp = MLPClassifier(hidden_layer_sizes=(5, 5), max_iter=500, tol=0.0001)
#Fit the data to the model
mlp.fit(X_train, y_train)
```

```python
y_pred = mlp.predict(X_train)
print(accuracy_score(y_pred, y_train))
y_pred_test = mlp.predict(X_test)
print(accuracy_score(y_pred_test, y_test))
```

### The accuracy for both train and test data is 47%
